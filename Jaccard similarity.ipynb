```python
import pandas as pd

# read amazon
df = pd.read_csv("amazon.csv")

# Grouping and aggregation
aggamazon = df.groupby("amazon", as_index=False).agg({"Date": "min"})

# result back to dataframe
amazon = pd.merge(df[["amazon", "Date"]], aggamazon, on=["amazon", "Date"], how="inner")

print(amazon)
```

```python
def jaccard_similarity(str1, str2):
    set1 = set(str1.split())
    set2 = set(str2.split())
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union)
```

```python
# Load the shein DataFrames from the csv files

shein = pd.read_csv("shein.csv")
amazon = pd.read_csv("amazon.csv")
```

```python
amazon_filter = (amazon['Date'] == '2023/2/9')
amazon = amazon.loc[amazon_filter]
amazon
```

```python
results = pd.DataFrame(columns=['Amazon Date',"Amazon Description","Amazon Rank",'Shein Date',"Shein Description","Shein Rank", "Jaccard Similarity"])
```

```python
# Iterate through each description in the Amazon DataFrame
for i, row_amazon in amazon.iterrows():
    # Iterate through each description in the Shein DataFrame
    for j, row_shein in shein.iterrows():
        similarity = jaccard_similarity(row_amazon['amazon'], row_shein["Description"])
        results = results.append({"Amazon Date": row_amazon["Date"], 
                                  "Amazon Description": row_amazon["amazon"],
                                  "Amazon Rank": row_amazon['Rank'],
                                  "Shein Date": row_shein['Date'],
                                  "Shein Description": row_shein["Description"],
                                  "Shein Rank": row_shein['Rank'],
                                  "Jaccard Similarity": similarity}, ignore_index=True)
        print(f"The {i}th row of amazon is computed! ")
```

```python
# Save the results to a new csv file
results.to_csv("results2.9.csv", index=False)
```

```python
df_max_items = pd.DataFrame()

list_items = results['Amazon Description'].unique()

for item in list_items:
    
    this_item_comparison = results[results['Amazon Description'] == item]
    
    this_item_comparison = this_item_comparison.sort_values(by = 'Jaccard Similarity', ascending = False).reset_index().drop('index',axis =1)
    
    frames = [df_max_items, this_item_comparison.loc[0, :].to_frame().T]
    
    df_max_items = pd.concat(frames)
```
