```python
# INSTALLING LIBRARIES (PANDAS)

import pandas as pd
amazon = pd.read_csv("Amazon Rankings 2023-01.csv")
amazon
```

```python
# WE NEED ONLY 5 VARIABLES FOR OUR ANALYSIS 

select_list = amazon[["Type","Category","Date","Rank","Title"]]
select_list
```

```python
# DATA NEEDS TO BE FILTERED FURTHER TO KITCHENWARE CATEGORY OF BESTSELLERS
# A filter is created to retrieve the kitchenware list

type_filter = (select_list['Type'] == 'Best Sellers' ) & (select_list['Category'] == 'Kitchen & Dining')
select_list[type_filter]
```

```python
select_list['Date'] = pd.to_datetime(select_list['Date'])
```

```python
# Created a date filter to record observations only from 23th January for the analysis

date_filter = select_list[select_list['Date'] >= '2023-01-23T00:00:00.000000000']
filter_amazon = date_filter[type_filter]
filter_amazon
```

```python
# convert dataframe to csv format

filter_amazon.to_csv("filter_amazon.csv", index=False)
```

```python
# reading new dataset in csv format

import pandas as pd
shein = pd.read_csv("shein.csv")
shein
```
